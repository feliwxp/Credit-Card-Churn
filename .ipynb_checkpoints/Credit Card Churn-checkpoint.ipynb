## Data Wrangling

```python
df['Gender'].replace({'M':0,'F': 1}, inplace = True)
```

```python
df = pd.concat([df,pd.get_dummies(df['Education_Level']).drop(columns=['Unknown'])],axis=1)
df = pd.concat([df,pd.get_dummies(df['Income_Category']).drop(columns=['Unknown'])],axis=1)
df = pd.concat([df,pd.get_dummies(df['Marital_Status']).drop(columns=['Unknown'])],axis=1)
df = pd.concat([df,pd.get_dummies(df['Card_Category']).drop(columns=['Platinum'])],axis=1)
df.drop(columns = ['Education_Level','Income_Category','Marital_Status','Card_Category','CLIENTNUM'],inplace=True)
```

```python
X = df.drop(['Attrition_Flag'],axis=1)
y = df['Attrition_Flag']
```

There is an imbalance of data. Only 16% of the data represents the churn customers. Let's do oversampling to balance the data.

```python
ros = RandomOverSampler(random_state=42)

# fit predictor and target variable

X_ros, y_ros = ros.fit_resample(X, y)

print('Original dataset shape', Counter(y))
print('Resample dataset shape', Counter(y_ros))
X_ros.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_ros,y_ros,shuffle=True,stratify = y_ros,test_size=0.3,random_state=2)
```

## Machine Learning

### Logistic Regression

```python
parameters_lg = {'penalty': ['none','l2','l1'], 'C': [0.001,0.01,0.1,1,10,100,1000], 'solver': ['newton-cg', 'lbfgs', 'liblinear']}

grid_lg = GridSearchCV(LogisticRegression(), param_grid = parameters_lg, 
                          cv = 3, n_jobs = -1, verbose = 2)

best_lg = grid_lg.fit(X_train, y_train)

best_lg.best_params_
```

```python
y_pred = grid_lg.predict(X_test)

y_pred_prob=grid_lg.predict_proba(X_test)[:,1]

print(classification_report(y_test, y_pred))

lg_score = accuracy_score(y_test,y_pred)
print(f"{round(lg_score*100,4)}% Accurate")

print('R^2 Training Score: {:.2f} \nR^2 Validation Score: {:.2f}'.format(grid_lg.score(X_train, y_train),
                                                                                             grid_lg.score(X_test, y_test)))
```

```python
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])

sns.heatmap(confusion_matrix, annot=True)
plt.title("confusion_matrix")
plt.show()
```

### Gradient Boosting Classifier

## Evaluation

Let's take a look some model insights from the random forest classifier model, which has the highest accuracy rate and a lowest false negative rate.

```python
features = X.columns
importances = rfc.feature_importances_
indices = np.argsort(importances)

plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

```python
features = pd.DataFrame(data=importances, index=X.columns, columns=['Importances'])
features.sort_values('Importances',ascending=False)
```

We can see that the top important feature is `Total_Trans_Am` and `Total_Trans_Ct`. Let's make our model simpler here by removing features with importances of less than or equal to 0.005.

```python
to_keep = features.index[features['Importances']>0.005]
```

```python
X_ros = X_ros[to_keep]
```

```python
X_ros = X_ros[to_keep]
X_train,X_test,y_train,y_test=train_test_split(X_ros,y_ros,shuffle=True,stratify = y_ros,test_size=0.3,random_state=2)
```

```python
rfc = RandomForestClassifier(bootstrap= True,
 max_depth= 15,
 max_features= 3,
 min_samples_leaf= 3,
 min_samples_split= 6,
 n_estimators= 200)

rfc.fit(X_train, y_train)

y_pred = rfc.predict(X_test)

y_pred_prob=rfc.predict_proba(X_test)[:,1]

print(classification_report(y_test, y_pred))

rfc_score = accuracy_score(y_test,y_pred)
print(f"{round(rfc_score*100,4)}% Accurate")

print('R^2 Training Score: {:.2f} \nR^2 Validation Score: {:.2f}'.format(rfc.score(X_train, y_train), 
                                                                                             rfc.score(X_test, y_test)))
```

```python
features = X_train.columns
importances = rfc.feature_importances_
indices = np.argsort(importances)

plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```
